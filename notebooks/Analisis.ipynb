## Librerias e importaci칩n de datos:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../data/sign_mnist_train.csv")
Xtest= data.iloc[:, 1:].values
ytest= data.iloc[:, :1].values.flatten()
```

```python
print"(min,Max) value of image :",(np.min(Xtest),np.max(Xtest))
```

```python
print"(min,Max) value of classes :",(np.min(ytest),np.max(ytest)),"\n",np.unique(ytest)
```

## Los datos:

```python
perm = np.random.permutation(range(Xtest.shape[0]))[0:50]
random_imgs   = Xtest[perm]
random_labels = ytest[perm] 
fig = plt.figure(figsize=(10,6))
for i in range(random_imgs.shape[0]):
    ax=fig.add_subplot(5,10,i+1)
    plt.imshow(random_imgs[i].reshape(28,28), interpolation="nearest")
    ax.set_title(int(random_labels[i]))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
```
